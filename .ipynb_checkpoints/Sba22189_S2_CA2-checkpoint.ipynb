```python
import pandas as pd
from itertools import combinations
from collections import Counter
from textblob import TextBlob
```

```python
batch_size = 500

# Initialize an empty Counter object
co_occurrences = Counter()

# Get the total number of rows in the dataframe
total_rows = len(DataFromDB)

# Iterate over the dataframe with defined batch size
for i in range(0, total_rows, batch_size):

    batch = DataFromDB['tweets'].iloc[i:i+batch_size]
    words = ' '.join(batch).split()

    # Count the co-occurring word pairs
    co_occurrences += Counter(list(combinations(words, 2)))


top_co_occurrences = co_occurrences.most_common()# Get the most common co-occurring word pairs

# Create a new dataframe for the words
CoupledWords = pd.DataFrame(top_co_occurrences, columns=['CoupledWords', 'Frequency'])

# Print
print(CoupledWords)
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
#nltk.download('vader_lexicon')

# Initialize the SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
def classify_sentiment(tweet):
    sentiment_scores = sia.polarity_scores(tweet)
    compound_score = sentiment_scores['compound']
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df_twitterdata['Sentiment'] = df_twitterdata['Tweet'].apply(classify_sentiment)
```

## One hot encoding

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping

```

```python

# Apply one-hot encoding to the "Sentiment" column
one_hot_encoded = pd.get_dummies(df_twitterdata['Sentiment'])

# Concatenate the one-hot encoded columns with the original DataFrame
df_encoded = pd.concat([df_twitterdata, one_hot_encoded], axis=1)
print(df_encoded)
```

```python
#pip install -U scikit-learn
```

```python
df = df_encoded[['Tweet', 'Positive', 'Negative', 'Neutral']].copy()
```

```python
X = df['Tweet'].values
y = df.drop('Tweet', axis=1).values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
vect = CountVectorizer()
X_train = vect.fit_transform(X_train)
X_test = vect.transform(X_test)
```

```python
tfidf = TfidfTransformer()
X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
X_train = X_train.toarray()
X_test = X_test.toarray()
```

```python
model = Sequential()

model.add(Dense(units=12673,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=4000,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=500,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=3, activation='softmax'))

opt=tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])

#early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=2)
```

```python
model.fit(x=X_train, y=y_train, batch_size=256, epochs=10, validation_data=(X_test, y_test), verbose=1)
```

```python
model_score = model.evaluate(X_test, y_test, batch_size=64, verbose=1)
print('Test accuracy:', model_score[1])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet

# Count the number of tweets per day
df_tweets = df_twitterdata[['UTC Date']].copy()
df_tweets['Tweet Count'] = 1
df_tweets = df_tweets.resample('D', on='UTC Date').sum().reset_index()

# Prepare the data
data = df_tweets[['UTC Date', 'Tweet Count']].copy()
data.columns = ['ds', 'y']  # Rename the columns for Prophet

model = Prophet()
model.fit(data)

# Make future dataframes for predictions
future_week = model.make_future_dataframe(periods=7)  # Next week
future_month = model.make_future_dataframe(periods=30)  # Next month
future_3months = model.make_future_dataframe(periods=90)  # Next 3 months

# Generate predictions for each future dataframe
forecast_week = model.predict(future_week)
forecast_month = model.predict(future_month)
forecast_3months = model.predict(future_3months)

# Plot the forecasts
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Plot for next week
axs[0].plot(forecast_week['ds'], forecast_week['yhat'], label='Forecast')
axs[0].scatter(data['ds'], data['y'], color='red', label='Observed')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Tweet Count')
axs[0].set_title('Tweet Count Forecast - Next Week')
axs[0].legend(loc='upper left')

# Plot for next month
axs[1].plot(forecast_month['ds'], forecast_month['yhat'], label='Forecast')
axs[1].scatter(data['ds'], data['y'], color='red', label='Observed')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Tweet Count')
axs[1].set_title('Tweet Count Forecast - Next Month')
axs[1].legend(loc='upper left')

# Plot for next 3 months
axs[2].plot(forecast_3months['ds'], forecast_3months['yhat'], label='Forecast')
axs[2].scatter(data['ds'], data['y'], color='red', label='Observed')
axs[2].set_xlabel('Date')
axs[2].set_ylabel('Tweet Count')
axs[2].set_title('Tweet Count Forecast - Next 3 Months')
axs[2].legend(loc='upper left')

# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
```

```python
# Count the number of tweets per day
df_tweets = df_twitterdata[['UTC Date']].copy()
df_tweets['Tweet Count'] = 1
df_tweets = df_tweets.resample('D', on='UTC Date').sum().reset_index()

# Set the 'UTC Date' column as the index
df_tweets.set_index('UTC Date', inplace=True)

# Split the data into training and testing sets
train_data = df_tweets.loc[:'2023-04-30']
test_data = df_tweets.loc['2023-05-01':]
```

```python
# Perform Dickey-Fuller test:
print('Results of Dickey-Fuller Test:')
dftest = adfuller(df_tweets['Tweet Count'], autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value

print(dfoutput)
```

```python
# Define the ARIMA model
model = ARIMA(train_data, order=(1, 1, 1))

# Fit the model to the training data
model_fit = model.fit()
```

```python
# Number of periods to forecast
forecast_periods = len(test_data)

# Generate predictions
predictions = model_fit.predict(start=len(train_data), end=len(train_data) + forecast_periods - 1)
```

```python
# Calculate the mean squared error
mse = ((predictions - test_data['Tweet Count'].values) ** 2).mean()

# Print the mean squared error
print(f"Mean Squared Error: {mse}")
```

```python
#figure size
plt.figure(figsize=(12, 6)) 


# Combine the training and testing data
combined_data = pd.concat([train_data, test_data])


plt.plot(combined_data.index, combined_data.values, label='Observed')# Plot the observed data


plt.plot(test_data.index, predictions, label='Forecast')# Plot the forecasted values

# Set the x-axis label and title
plt.xlabel('Date')
plt.ylabel('Tweet Count')
plt.title('ARIMA Tweet Count Forecast')

plt.legend()
plt.show()
```
